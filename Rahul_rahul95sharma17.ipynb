## predictive model to assess whether a loan application will be approved or rejected and provide the accuracy score

```python
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
# Converting data into binary values
column=['Gender','Married','Dependents','Education','Self_Employed','Credit_History','Property_Area','Income','Application_Status']
all= LabelEncoder()
for i in column:
    df[i] = all.fit_transform(df[i])
    
df
```

```python
# Converting Data in Train and Test Set
x=df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed','Credit_History','Property_Area', 'Income']]
y=df.Application_Status
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
#Logistic Regression
model = LogisticRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)
print(accuracy_score(y_test, predictions)*100)
```

```python
#Using Logistic Regression Algorithm to the Training Set
from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state = 0)
classifier1.fit(x_train, y_train)
#GaussianNB method
from sklearn.naive_bayes import GaussianNB
classifier2 = GaussianNB()
classifier2.fit(x_train, y_train)
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
classifier3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier3.fit(x_train, y_train)
```

```python
Y_pred1 = classifier1.predict(x_test)
Y_pred2 = classifier2.predict(x_test)
Y_pred3 = classifier3.predict(x_test)
```

```python
print(accuracy_score(y_test, Y_pred1))
print(accuracy_score(y_test, Y_pred2))
print(accuracy_score(y_test, Y_pred3))
```
